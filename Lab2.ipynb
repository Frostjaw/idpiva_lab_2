```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Детектор Харриса

```python
img = cv2.imread('image_1.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

#### Применение детектора

```python
hcd_img = cv2.cornerHarris(gray, 5, 5, 0.08)
```

#### Поиск углов

```python
hcd_img = cv2.dilate(hcd_img, None)
```

#### Выделение углов на исходном изображение

```python
img[hcd_img > 0.01 * hcd_img.max()]=[255, 0, 0] 
```

#### Отображение графики

```python
fig = plt.figure(figsize=(20, 7))
rows = 1
columns = 2

titles = [u'Оригинальное изображение ', u'Детектор Харриса']
images = [cv2.imread('image_1.png'), img]
for i in range(2):
    fig.add_subplot(rows, columns, i + 1)
    plt.imshow(images[i] , 'gray')
    plt.axis('off')
    plt.title(titles[i])
plt.show()
```

### Выделение ключевых точек SIFT

```python
img = cv2.imread('image_2.jpg')  
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

#### Применение SIFT

```python
sift = cv2.xfeatures2d.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(img, None)
img = cv2.drawKeypoints(gray, keypoints, img)
```

#### Отображение графики

```python
fig = plt.figure(figsize=(20, 7))
rows = 1
columns = 2

titles = [u'Оригинальное изображение ', u'Выделение ключевых точек SIFT']
images = [cv2.imread('image_2.jpg'), img]
for i in range(2):
    fig.add_subplot(rows, columns, i + 1)
    plt.imshow(images[i])
    plt.axis('off')
    plt.title(titles[i])
plt.show()
```

### Выделение ключевых точек HOG

```python
import skimage
from skimage.transform import resize
from skimage.feature import hog
```

```python
img = cv2.imread('image_2.jpg')
```

#### Изменение размера изображения

```python
resized_img = resize(img, (128*4, 64*4))
```

#### Применение HOG

```python
fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
```

#### Отображение графики

```python
fig = plt.figure(figsize=(20, 7))
rows = 1
columns = 2

titles = [u'Оригинальное изображение ', u'Выделение ключевых точек HOG']
images = [cv2.imread('image_2.jpg'), hog_image]
for i in range(2):
    fig.add_subplot(rows, columns, i + 1)
    plt.imshow(images[i])
    plt.axis('off')
    plt.title(titles[i])
plt.show()
```
